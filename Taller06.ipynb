# Euler


```python
from src import ODE_euler
import numpy as np

import matplotlib.pyplot as plt

def f(t, y):
    return y - t**2 + 1

a = 0
y0 = 0.5
b = 2
N = 10

y_values, t_values, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=N)

print("t_values\t y_values")
print("------------------------")
for t, y in zip(t_values, y_values):
    print(f"{t:.4f}\t {y:.4f}")

def exact_solution(t):
    return (t + 1)**2 - 0.5 * np.exp(t)

t_exact = np.linspace(a, b, 100)
y_exact = exact_solution(t_exact)
```

```python
y_values2, t_values2, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=5)

print("t_values\t y_values")
print("------------------------")
for t, y in zip(t_values2, y_values2):
    print(f"{t:.4f}\t {y:.4f}")
```

```python
y_values20, t_values20, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=20)

print("t_values\t y_values")
print("------------------------")
for t, y in zip(t_values20, y_values20):
    print(f"{t:.4f}\t {y:.4f}")
```

```python
y_values15, t_values15, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=15)

print("t_values\t y_values")
print("------------------------")
for t, y in zip(t_values15, y_values15):
    print(f"{t:.4f}\t {y:.4f}")
```

```python
plt.plot(t_values, y_values, label='Euler con N=10', marker='o')
plt.plot(t_values2, y_values2, label='Euler con N=5', marker='o')
plt.plot(t_values20, y_values20, label='Euler con N=20', marker='o')
plt.plot(t_values15, y_values15, label='Euler con N=15', marker='o')
plt.plot(t_exact, y_exact, label=' Solucion Exacta', linestyle='--')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solución ODE usando el método de Euler')
plt.legend()
plt.grid()
plt.show()
```

```python
from src import ODE_euler
import numpy as np

import matplotlib.pyplot as plt

def f(t, y):
    return y/t - (y/t)**2

a = 1
y0 = 1
b = 2
N = 10

y_values, t_values, h = ODE_euler(a=a, b=b, f=f, y_t0=y0, N=N)

print("t_values\t y_values")
print("------------------------")
for t, y in zip(t_values, y_values):
    print(f"{t:.4f}\t {y:.4f}")

def exact_solution(t):
    return t/(1 + np.log(t))

t_exact = np.linspace(a, b, 100)
y_exact = exact_solution(t_exact)
```
